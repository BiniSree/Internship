```python
## Removing outliers of Monthly Inhand Salary
```

```python
data1= x['Monthly_Inhand_Salary']
Q1mns= data1.quantile(0.25)
Q3mns= data1.quantile(0.75)
IQRmns= Q3mns-Q1mns
IQRmns
```

```python
threshold= 1
outliermns= x[(data1 < (Q1mns-(1*IQRmns))) | (data1 > (Q3mns+(1*IQRmns)))]
outliermns
```

```python
x['Monthly_Inhand_Salary'].mean()
mean_mns= x['Monthly_Inhand_Salary'].mean()
mean_mns
```

```python
outliermns.values
```

```python
outliermns['Monthly_Inhand_Salary'].values
```

```python
x= x.replace(to_replace= outliermns['Monthly_Inhand_Salary'].values, value= mean_mns, inplace=False)
```
