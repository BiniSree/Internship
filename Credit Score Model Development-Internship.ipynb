## Exploratory Data Analysis

```python
age= data['Age']
age.value_counts().plot(kind= 'bar', figsize= (8,5))
plt.xlabel= 'Age'
plt.ylabel= 'Count'
plt.title= 'Countplot of Age Category of Credit Availers'
plt.show()
```

```python
plt.figure(figsize= (10,8))
sns.scatterplot(y= 'Credit_Utilization_Ratio', x= 'Monthly_Inhand_Salary', data= data, s=10)
plt.xlabel= 'Credit Utilization Rate'
plt.ylabel= 'Monthly Inhand Salary'
plt.title= 'Relation of Credit Utilization Rate and Monthly Inhand Salary'
plt.show()
```

```python
plt.figure(figsize= (8,6))
sns.lineplot(x= 'Occupation', y= 'Monthly_Balance', data= data)
plt.xlabel= 'Occupation'
plt.xticks(rotation=45)
plt.ylabel= 'Monthly_Balance'
plt.title= 'Line plot of Occupation and Monthly Balance'
plt.show()
```

```python
plt.figure(figsize= (10,5))
sns.violinplot(x= 'Occupation', y= 'Num_of_Delayed_Payment', data= data)
plt.xlabel= 'Occupation'
plt.xticks(rotation=45)
plt.ylabel= 'No:of Delayed Payments'
plt.title= 'Relation of occupation and Delayed Payments grouped by No: of loans'
plt.show()
```

```python
plt.figure(figsize= (10,8))
sns.boxplot(x= 'Num_of_Loan', y= 'Outstanding_Debt', hue= 'Credit_Score', data= data)
plt.xlabel= 'No: of Loans'
plt.ylabel= 'Outstanding Debt'
plt.title= 'Relation between No:of Loans and Outstanding Debt grouped by Credit Score'
plt.show()
```

```python
sns.barplot(x= 'Credit_Mix', y= 'Outstanding_Debt', data= data)
plt.xlabel= 'Credit Mix Category'
plt.ylabel= 'Outstanding Debt'
plt.title= 'Barplot of Credit Mix Vs Outstanding Debt'
plt.show()
```

```python
sns.barplot(y= 'Annual_Income', x= 'Payment_of_Min_Amount', data= data)
plt.xlabel= 'Annual Income'
plt.ylabel= 'Payment of Min. Amount'
plt.title= 'Scatterplot of Annual Income and Payment of Min. Amount'
plt.show()
```

```python
sns.lineplot(x= 'Payment_Behaviour', y= 'Monthly_Inhand_Salary', data= data)
plt.xlabel= 'Payment Behaviour'
plt.xticks(rotation=45)
plt.ylabel= 'Monthly Inhand Salary'
plt.title= 'Monthly Salary Vs Payment Behaviour'
plt.show()
```

```python
sns.scatterplot(x= 'Interest_Rate', y= 'Outstanding_Debt', data= data)
plt.xlabel= 'Interest Rate'
plt.ylabel= 'Outstanding Debt'
plt.title= ''
plt.show()
```

```python
sns.barplot(x= 'Changed_Credit_Limit', y= 'Credit_Score', data= data)
plt.xlabel= 'Changed Credit Limit'
plt.ylabel= 'Credit Score'
plt.show()
```
